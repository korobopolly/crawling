# 파이썬 기상청 api 크롤링 실습

```python
#c4

from urllib import request

api="http://www.kma.go.kr/wid/queryDFSRSS.jsp"
param="zone=4117354000" #zone값을 매개변수로 설정
url=api+"?"+param
data=request.urlopen(url).read()
text=data.decode("utf-8") #byte형의 데이터를 문자열로 디코딩
print(text)
```

```python
#c5 뷰티풀수프 이용

from bs4 import BeautifulSoup

api="http://www.kma.go.kr/wid/queryDFSRSS.jsp"
param="zone=4117354000" #zone값을 매개변수로 설정
url=api+"?"+param
html=request.urlopen(url).read()
soup=BeautifulSoup(html,'html.parser')
print(soup)
```

```python
#c6

all_datas=soup.find("data")
print(all_datas)
```

```python
#c7

all_datas=soup.find_all("data")
print(all_datas)
```

```python
#c8

data_15=soup.find_all("data",{"seq":"15"})
print(data_15)
```

```python
#c9

for i, data in enumerate(all_datas):
    print("{0}번째".format(i+1), data, "끝", sep="\n")
```

```python
#c10

for i, data in enumerate(all_datas):
    day_data=soup.find("data",{"seq":i})
    day=day_data.find("day")
    time=day_data.find("hour")
    temp=day_data.find("temp")
    #print(day_data)
    print("일 : ",day.string)
    print("시각 : ",time.string)
    print("기온 : ",temp.string)
    print()
```
